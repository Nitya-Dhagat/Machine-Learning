# **Salary predictor**
## *Linear Regression Project 3*

This is ML model that predicts Salary the a candidate should deserve based upon his Experience, test_score and Interview_score.
This is example of Multivarite Linear regression model.

The csv file is really small with columns :

Index(['experience', 'test_score(out of 10)', 'interview_score(out of 10)',
       'salary($)'],
      dtype='object')

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from word2number import w2n
from sklearn import linear_model
```

```python
minimum_marks = 3
maximum_marks = 10
```

Import the csv file and read it contents, in order to understand what preprocessing is required.
We can observe that we have null values in experience as well as test_score.
We will replace experience null values with zero and test_score null values with minimum_marks that is 4.

```python
df = pd.read_csv("hiring.csv")
df
```

Firstly, we will handle NaN values in the experience section by replacing it with Zero.

```python
df.experience = df.experience.fillna("zero")
df
```

Now change the NaN values of test_score with minimum_marks.

```python
df = df.rename(columns={"test_score(out of 10)": "test_score"})
df = df.rename(columns={"interview_score(out of 10)": "interview_score"})
df.test_score = df.test_score.fillna(minimum_marks)
df
```

 Convert the experience field values from string to int using word2nuber library of Python.

```python
df.experience = df.experience.apply(w2n.word_to_num)
df
```

```python
reg_model = linear_model.LinearRegression()
reg_model.fit(df[['experience', 'test_score', 'interview_score']], df['salary($)'])
```

```python
print(reg_model.coef_)
print(reg_model.intercept_)
```

```python
reg_model.predict([[2, 9, 6]])
reg_model.predict([[12, 10, 10]])
```

This model has achieved a accuracy of 98.40%

```python
print(reg_model.score(df[['experience', 'test_score', 'interview_score']], df['salary($)']))
```
