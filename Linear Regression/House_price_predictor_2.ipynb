#**House Price Predicator**
##*Linear Regression Project 2*

This is ML model that predicts house price the a candidate should deserve based upon the area of the house. This is example of Linear regression model.

I have utilised kaggle daaset for this purposes. Link to the dataset : https://www.kaggle.com/code/ammar111/house-price-prediction-an-end-to-end-ml-project

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

df = pd.read_csv('/content/sample_data/california_housing_train.csv')
df.head()
```

Using data visualisation, we will look the trends of Age vs Income for the given dataset.

```python
age = np.array(df['housing_median_age']).reshape(-1,1)
income = np.array(df['median_income']).reshape(-1,1)

plt.scatter(age, income,color='blue',marker='+')
plt.show()

plt.hist(age,bins=100)
plt.hist(income,bins=100)
plt.show()
```

```python
model = LinearRegression()
model.fit(age,income)
```

```python
model.predict([[21]])
model.coef_
model.intercept_
```

```python
df_test = pd.read_csv('/content/sample_data/california_housing_test.csv')
df_test.head()
```

```python
age_test = np.array(df_test['housing_median_age']).reshape(-1,1)
income_test = np.array(df_test['median_income']).reshape(-1,1)
y_pred = model.predict(age_test)
plt.scatter(age_test, income_test,color='blue',marker='+')
plt.plot(age_test, y_pred,color='red')
plt.show()
```

```python
print(model.score(age_test, income_test))
```
