# **Hand-written digit predicator**
## *Logistic Regression Project 1*

This is a ML model that is used to predict hand-written digits. The dataset is provided by the sklearn library.
This is classification problem, hence best approve is to use Logistic Regression.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
```

```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
```

This shows different directories present in dataset.
data = It has the value of each cell of 8x8 matrix corresponding to a number.
images = It is a visual representation how the hand-written digit looks in 8x8 matrix,
target = It contains the actual number in decimal which we are trying to represent using image.

```python
digits = load_digits()
dir(digits)
```

```python
for i in range(5):
  plt.matshow(digits.images[i])
```

```python
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.2)
```

Applying Linear Regression gives a poor performance of only 56.6%, which is not efficient, because it is not a predication probelm.

```python
model_reg = LinearRegression()
model_reg.fit(X_train, y_train)
model_reg.predict(X_test)
model_reg.score(X_test,y_test)
```

```python
model_log = LogisticRegression()
model_log.fit(X_train, y_train)
model_log.predict(X_test)
```

Logistic Regression gives a great accuracy of 98.33%

```python
model_log.score(X_test,y_test)
```

```python
y_pred  = model_log.predict(X_test)
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, color=sns.color_palette("Blues"))
```
