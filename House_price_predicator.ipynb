# **House price predictor**
## *Linear Regression Project 1*

This is ML model that predicts house price the a candidate should deserve based upon the area of the house.
This is example of Linear regression model.

I have utilised kaggle daaset for this purposes. Link to the dataset : https://www.kaggle.com/code/ammar111/house-price-prediction-an-end-to-end-ml-project


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import *

df = pd.read_csv('train.csv')
df.head()
```

```python
%matplotlib inline
plt.xlabel('sq. ft')
plt.ylabel('price')
plt.scatter(df.LotArea,df.SalePrice, color = 'red', marker = '+')
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df[['LotArea']],df.SalePrice, test_size = 0.2)
```

```python
reg = linear_model.LinearRegression()
reg.fit(X_train,y_train)
reg.predict([[3000]])
```

```python
reg.coef_
reg.intercept_
ans = reg.coef_ * 3000 + reg.intercept_
print(ans)
```

```python
test_score = reg.score(X_test,y_test)
print(test_score)
y_pred = reg.predict(X_test)
print(y_pred)
```

```python
%matplotlib inline
plt.xlabel('sq. ft', size=20)
plt.ylabel('price',size=20)
plt.scatter(df.LotArea,df.SalePrice, color = 'red', marker = '+')
plt.plot(X_test,y_pred,color = 'blue')
plt.show()
```
